```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis, y_axis, 'o')
```

```python
#plt.scatter(x_axis, y_axis, s=y_axis, color= "red", label ="Chicago")

plt.plot(y_axis, x_axis, 'o', color= "red", label ="Chicago")
plt.gca().invert_yaxis()

# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare ($)")


# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()


# Add a grid.
plt.grid()
```

```python
# Make the size of markers 3 times bigger depending on size of Y axis

##y_axis_larger = []
##for data in y_axis:
    ##y_axis_larger.append(data*3)
##plt.scatter(x_axis, y_axis, s=y_axis_larger, color= "green", label ="Chicago")

# Alternate way of makign markers bigger using list comprehension without a for loop:

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis], color= "skyblue", edgecolor= "black", label ="Chicago")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")


# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
categories = ['pants', 'shirts', 'outerwear', 'pajamas', 'underwear', 'shoes']
revenue = [25,15,5,20,10,25]
plt.pie(revenue, labels = categories, autopct="%1.1f%%" )
plt.show()
```

```python
explode = [0,.1,0,0,0,0]
```

```python
plt.pie(revenue, labels = categories, autopct="%1.1f%%",explode=explode )
plt.show()
```

```python
plt.subplots(figsize=(10, 10))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "mistyrose", "indianred", "tomato", "lightgrey"]

# Show july performancer separately using explode
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Increase pie-chart size to make it look clean
plt.subplots(figsize=(8, 8))

plt.pie(y_axis,
    labels=x_axis,
    colors=colors,
    explode=explode_values,
    autopct='%.1f%%',
    shadow=True,
    startangle=140,
    'radius = 1.2)

plt.show()
```
