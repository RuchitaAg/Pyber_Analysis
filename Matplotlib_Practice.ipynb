```python
%matplotlib inline
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Recreate the line chart using the object-oriented interface method; ax.plt()
fig = plt.axes()
fig.plot(x_axis, y_axis)
```

```python
plt.show()
```

```python
# Create the plot using MATLAB
plt.plot(x_axis, y_axis, label = 'Boston')
plt.xlabel('Months')
plt.ylabel('Total Fare; $US')
plt.title("PyBer Fare by Month")
#plt.grid()
plt.savefig("./Total Fare By Month Line Chart")
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="red", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Set the x limit between Jan and June.
plt.xlim("June", "Dec")

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()

# change the figure size to 20 pixels wide by 10 pixels high
plt.figure(figsize=(20, 10))
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Creating Bar Charts:

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot
plt.bar(x_axis, y_axis, color = "LightGreen",  label = "Boston")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Set the x limit between Jan and June.
plt.xlim("Jan", "Dec")

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
#plt.grid()

# Add the legend.
plt.legend()
```

```python
# Plot horizontal bar graph

plt.barh(x_axis, y_axis, color = "lightGreen",  label = "Boston")
```

```python
# Plot horizontal bar graph with Y axis in inverse order Jan-Dec

plt.barh(x_axis, y_axis, color = "Magenta",  label = "Boston")

# gca = get curretn axis
plt.gca().invert_yaxis()

# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare ($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
#plt.grid()

# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color = "lightblue", label = "New York", yerr=stdev, capsize=3)
#ax.set_color('brown')
ax.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label = "New York")
plt.gca().invert_yaxis()
#ax.set_color('brown')
ax.legend()
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.set_title("PyBer Fare by Month")
```

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# Add minor ticks without numbers

from matplotlib.ticker import MultipleLocator

# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
